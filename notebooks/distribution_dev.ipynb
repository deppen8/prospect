```python
binom_vals = binom.rvs(n=1, p=0.8, size=10000);
ax = sns.distplot(binom_vals,
                  bins=20,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Binomial Distribution ', ylabel='Frequency');
```

```python
beta_vals = beta.rvs(a=9, b=1, size=10000);
ax = sns.distplot(beta_vals,
                  bins=20,
                  kde=True,
#                   color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Beta Distribution ', ylabel='Frequency');
```

```python
norm_vals = norm.rvs(size=10000);
ax = sns.distplot(norm_vals,
                  bins=20,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Normal Distribution ', ylabel='Frequency');
```

```python
truncnorm_vals = truncnorm.rvs(a=0, b=100, loc=10, scale=10, size=10000);
ax = sns.distplot(truncnorm_vals,
                  bins=20,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Truncated Normal Distribution ', ylabel='Frequency');
```

```python
halfnorm_vals = halfnorm.rvs(loc=0, scale=, size=10000);
ax = sns.distplot(halfnorm_vals,
                  bins=20,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Truncated Normal Distribution ', ylabel='Frequency');
```

```python
def make_truncnorm_distribution(mean: float, sd: float, lower: float = 0.0, upper: float = 999999) -> rv_continuous:
    return truncnorm((lower - mean) / sd, (upper - mean) / sd, loc=mean, scale=sd)
tn = make_truncnorm_distribution(mean=10, sd=2)
truncnorm_vals = tn.rvs(size=10000);
ax = sns.distplot(truncnorm_vals,
                  bins=20,
                  kde=True,
                  color='skyblue',
                  hist_kws={"linewidth": 15,'alpha':1})
ax.set(xlabel='Truncated Normal Distribution ', ylabel='Frequency');
```
