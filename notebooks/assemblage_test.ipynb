```python
from surveysim.assemblage import Assemblage
from surveysim.layer import Layer
from surveysim.area import Area
%matplotlib inline
```

```python
a = Area.from_area_value('area1', 100)
```

```python
layer_lst = []
for i in range(10):
    layer = Layer.from_pseudorandom_points(i, a, f'layer_{i}')
    layer_lst.append(layer)
```

```python
layer_lst[0].data.plot()
```

```python
assem = Assemblage('10layers', layer_lst)
```
